```python
#creating variable:

Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Yearz=[]

time.sleep(2)

#fetching Dataset_name:
try:
    Dataset_tag=driver.find_elements_by_xpath("//p[@class='normal']")
    for i in Dataset_tag:
        datasetz=i.text
        Dataset_name.append(datasetz)
except NoSuchElementException as e:
    Dataset_name.append('-')
    time.sleep(2)
    
#fetching Data_type:
try:
    data_tag=driver.find_elements_by_xpath("//p[@class='normal']")
    for i in data_tag:
        datas=i.text
        Data_type.append(datas)
except NoSuchElementException as e:
    Data_type.append('-')

#fetching Task:
try:
    Task_tag=driver.find_elements_by_xpath("//p[@class='normal']")
    for i in Task_tag:
        tasks=i.text
        Task.append(tasks)
except NoSuchElementException as e:
    Task.append('-')

#fetching Attribute_type:
try:
    Attribute_tag=driver.find_elements_by_xpath("//p[@class='normal']")
    for i in Attribute_tag:
        att=i.text
        Attribute_type.append(att)
except NoSuchElementException as e:
    Attribute_type.append('-')
    
#fetching No_of_instances:
try:
    instances_tag=driver.find_elements_by_xpath("//p[@class='normal']")
    for i in instances_tag:
        instances=i.text
        No_of_instances.append(instances)
except NoSuchElementException as e:
    No_of_instances.append('-')
    
#fetching No_of_attribute:
try:
    na_tag=driver.find_elements_by_xpath("//p[@class='normal']")
    for i in na_tag:
        noa=i.text
        No_of_attribute.append(noa)
except NoSuchElementException as e:
    No_of_attribute.append('-')
    
#fetching Yearz:
try:
    Yearz_tag=driver.find_elements_by_xpath("//p[@class='normal']")
    for i in Yearz_tag:
        Yearzs=i.text
        Yearz.append(Yearzs)
except NoSuchElementException as e:
    Yearz.append('-')
```
