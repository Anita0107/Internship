Q 1. Write a python program which searches all the product under a particular product vertical from www.amazon.in. The product verticals to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.
Q2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
#url = "https://www.amazon.in/"
driver.get("https://www.amazon.in/")
time.sleep(2)

#taking user input
a=input()
search_bar=driver.find_element_by_id('twotabsearchtextbox')
search_bar.send_keys(a) 

# clicking the search button
search_btn=driver.find_element_by_id('nav-search-submit-button')
search_btn.click()
time.sleep(2)
```

```python
#fetching url link of each element
urls=[]
for i in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
    urls.append(i.get_attribute("href"))     
```

```python
#defining variables
url_link=[]
brandname=[]
productname=[]
ratings=[]
numofratings=[]
prices=[]
return_exch=[]
expected_deli=[]
availability=[]
product_u=[]
time.sleep(3)

#scrapping the required details
page_tag = driver.find_element_by_xpath("//li[@class='a-last']/a")
for i in range(1,4):
    url_link.append(page_tag.get_attribute('href')[0:-1]+str(i))

for i in url_link:
    driver.get(i)
    time.sleep(2)
    for i in urls:
        driver.get(i)
        time.sleep(3)
        
        #fetching brand name:
        try:
            brands=driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']") 
            brand=brands.text.split(' ')
            brandname.append(brand)
        except NoSuchElementException:
            brandname.append('-')
        
        #fetching product name
        try:
            pn=driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']")
            pnss=pn.text.split(' ')[1:]
            productname.append(pnss)
        except NoSuchElementException:
            productname.append('-')
        
        #fetching Ratings:
        try:                 
            Rating=driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']//span")
            ratings.append(Rating.text)
        
        except NoSuchElementException:
            ratings.append("NO rating")
        
        # fetching number of ratings
        try:
            numrating=driver.find_element_by_xpath("//div[@data-hook='total-review-count']/span")
            numofratings.append(numrating.text)
        except NoSuchElementException:
            numofratings.append('-')
            
        # fetching prices
        try:
            price_tags=driver.find_elements_by_xpath("//td[@class='a-span12']/span[1]")
            for i in price_tags:
                price=i.text
                prices.append(price)
        except NoSuchElementException:
            prices.append('-')
            
        # fetching return/exchange details
        try:
            re=driver.find_element_by_xpath("//div[@class='a-column a-span3 a-text-center icon-container icon-grid-width celwidget']/span/div/a")
            return_exch.append(re.text)#appending the text in Brand list
        except NoSuchElementException:
            return_exch.append('-')
            
        # fetching expected delivery
        try:
            ed=driver.find_element_by_xpath("//div[@class='a-section a-spacing-mini']/b")#collecting urls of all the laptop
            expected_deli.append(ed.text)#appending the text in Brand list
        except NoSuchElementException:
            expected_deli.append('-')
            
        # fetching availability of product
        try:
            ava=driver.find_element_by_xpath("//div[@class='a-section a-spacing-none }']/span")#collecting urls of all the laptop
            availability.append(ava.text)#appending the text in Brand list
        except NoSuchElementException:
            availability.append('-')
        #fetching product url    
        try:
            product_url=driver.current_url
            product_u.append(product_url)
        except NoSuchElementException:
            product_u.append('-')       
```

```python
df1=pd.DataFrame({'Brand': brandname,'product name': productname,
                 'Price': prices,'Ratings': ratings,'num of ratings': numofratings,
                 'Return/Exchange': return_exch,'Expected Delivery': expected_deli,'Availability': availability,
                 "Product url":product_u})
df1 
```

```python
#saving as csv file:
df1.to_csv('Amazon_data.csv', index=False, sep=';')
```

Q 8. Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
#Q8.

import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import time
import urllib.request
from bs4 import BeautifulSoup
from selenium.webdriver.common.action_chains import ActionChains

#creating variables
rich_info = []
page = []

rank = []
name = []
net_worth = []
age = []
citizenship = []
source = []
industry = []

PATH = 'C:\Program Files (x86)\chromedriver.exe' #extracting path of the webdriver
driver = webdriver.Chrome(r"C:\\Users\\hp\\Downloads\\chromedriver_win32\\chromedriver.exe")
driver.get("http://www.forbes.com/")
time.sleep(2)

search_btn = driver.find_element_by_xpath("//button[@class='icon--hamburger']")
search_btn.click()

time.sleep(1)
#fetching Billionaires data
a = ActionChains(driver)
m = driver.find_element_by_link_text("Billionaires")
a.move_to_element(m).perform()

m = driver.find_element_by_link_text("World's Billionaires")
a.move_to_element(m).perform()
m.click()

time.sleep(2)

for i in driver.find_elements_by_xpath("//button[@class='goto-page__btn']"):
    page.append(i.get_attribute("aria-label"))


for i in range(0,len(page)):
    
    info = driver.find_element_by_xpath("//button[@aria-label='{}']".format(page[i]))
    info.click()
    
    time.sleep(2)
    
    info = driver.find_element_by_xpath("//div[@class='table-row expanded']")
    rich_info.append(info.text.split('\n'))
    
    info = driver.find_elements_by_xpath("//div[@class='table-row ']")
    for i in range(len(info)):
        rich_info.append(info[i].text.split('\n'))
        
    
for i in range(0,len(rich_info)):
    rank.append(rich_info[i][0])
    name.append(rich_info[i][1])
    net_worth.append(rich_info[i][2])
    age.append(rich_info[i][3])
    citizenship.append(rich_info[i][4])
    source.append(rich_info[i][5])
    industry.append(rich_info[i][6])

#DataFrame:
billionaire_list = pd.DataFrame({})
billionaire_list['Rank'] = rank[0:]
billionaire_list['Name'] = name[0:]
billionaire_list['Net Worth'] = net_worth[0:]
billionaire_list['Age'] = age[0:]
billionaire_list['Citizenship'] = citizenship[0:]
billionaire_list['Source'] = source[0:]
billionaire_list['Industry'] = industry[0:]
```

```python
#saving as csv file:
billionaire_list.to_csv('World_Billionaires.csv', index=False, sep=';')
```

Q10. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
#Q 10.
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import time
import urllib.request
from bs4 import BeautifulSoup

#creating variables
property_info = []
price_info = []

hostel_name = []
distance = []
rating = []
total_review = []
overall_review = []

private = []
dorm = []

property_desc = []
facilities = []

urls = []

PATH = 'C:\Program Files (x86)\chromedriver.exe' #extracting path of the webdriver
driver = webdriver.Chrome(r"C:\\Users\\hp\\Downloads\\chromedriver_win32\\chromedriver.exe")
driver.get("https://www.hostelworld.com/")

time.sleep(2)

search_btn = driver.find_element_by_xpath("//button[@class='icon-core-menu-fill header-option item']")
search_btn.click()
time.sleep(1)
search_btn = driver.find_element_by_xpath("//a[@href='https://www.hostelworld.com/hostels']")
search_btn.click()

search_btn = driver.find_element_by_xpath("//input[@placeholder='search by city or hostel name']")
search_btn.send_keys('London')

time.sleep(1)

search_btn = driver.find_element_by_xpath("//button[@class='btn btn-primary btn-block']")
search_btn.click()

time.sleep(10)

info = driver.find_elements_by_xpath("//div[@class='property']")
for i in range(len(info)):
    property_info.append(info[i].text.split('\n'))
    
#print(property_info)
#print(len(property_info))

info = driver.find_elements_by_xpath("//div[@class='prices-col']")
for i in range(len(info)):
    price_info.append(info[i].text.split('\n'))
    
#print(price_info)
#print(len(price_info))

for i in driver.find_elements_by_xpath("//h2[@class='title title-6']//a[@class='']"):
    urls.append(i.get_attribute("href"))

for i in range(0,len(urls)):
    driver.get(urls[i])
    time.sleep(3)
    
    info = driver.find_element_by_xpath("//div[@class='content collapse-content']")
    property_desc.append(info.text.replace("\n",""))
    
    info = driver.find_elements_by_xpath("//ul[@class='facility-sections']")
    for i in range(len(info)):
        facilities.append(info[i].text.replace("\n"," "))

for x in range(0,30):
    hostel_name.append(property_info[x+3][0])
    distance.append(property_info[x+3][1].split(' Show on Map')[0])
    rating.append(property_info[x+3][-3])
    total_review.append(property_info[x+3][-2])
    overall_review.append(property_info[x+3][-1])
    
    private.append(price_info[x][-4])
    dorm.append(price_info[x][-2])
    
property_list = pd.DataFrame({})
property_list['Hostel Name'] = hostel_name[0:30]
property_list['Distance from city'] = distance[0:30]
property_list['Rating'] = rating[0:30]
property_list['Total Review'] = total_review[0:30]
property_list['Overall Review'] = overall_review[0:30]
property_list['Privates from Price'] = private[0:30]
property_list['Dorms from Price'] = dorm[0:30]
property_list['Facilities'] = facilities[0:30]
property_list['Property Description'] = property_desc[0:30]
```
