Q 4. Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. Product Description
3. Price

```python
import selenium
import pandas as pd
from selenium import webdriver
#from selenium.webdriver.common.keys import Keys
PATH = 'C:\Program Files (x86)\chromedriver.exe'                         
driver= webdriver.Chrome(r"C:\\Users\\hp\\Downloads\\chromedriver_win32\\chromedriver.exe")

import time

driver.get("https://www.flipkart.com/")

time.sleep(2)

search_btn=driver.find_element_by_xpath("//div[@class='_2QfC02']/button")
search_btn.click()

search_job=driver.find_element_by_class_name('_3704LK')
search_job.send_keys("sunglasses")

search_btn=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_btn.click()

brand=[]
prod_desc=[]
price=[]
perc_off=[]
url_link=[]

time.sleep(2)
#pagination 
page_tag = driver.find_element_by_xpath("//a[@class='_1LKTO3']")
for i in range(1,4):
    url_link.append(page_tag.get_attribute('href')[0:-1]+str(i))

for i in url_link:
    driver.get(i)
    time.sleep(2)
    
    #Brand name
    brand_tag = driver.find_elements_by_xpath("//div[@class='_2WkVRV']")  #Brand name
    for i in range(len(brand_tag)):
        brand.append(brand_tag[i].text)
   #Product name     
    prod_tag = driver.find_elements_by_xpath("//a[@class='IRpwTa']")      
    for i in range(len(prod_tag)):
        prod_desc.append(prod_tag[i].text)
   #Prices
    price_tag = driver.find_elements_by_xpath("//div[@class='_30jeq3']")  
    for i in range(len(price_tag)):
        price.append(price_tag[i].text)
    #percentage off
    per_tag = driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']/span[1]") 
    for i in range(len(per_tag)):
        perc_off.append(per_tag[i].text)

#DataFrame:
item_list = pd.DataFrame({})
item_list['Brand'] = brand[0:100]
item_list['Product Discription'] = prod_desc[0:100]
item_list['Price'] = price[0:100]
item_list['Percent Off'] = perc_off[0:100]
```

Q10. Write a python program to scrape the salary data for Data Scientist designation.
You have to scrape Company name, Number of salaries, Average salary, Min salary, Max Salary.

```python
import selenium
import pandas as pd
from selenium import webdriver
import time
#from selenium.webdriver.common.keys import Keys
PATH = 'C:\Program Files (x86)\chromedriver.exe'  
driver= webdriver.Chrome(r"C:\\Users\\hp\\Downloads\\chromedriver_win32\\chromedriver.exe")
driver.get("https://www.ambitionbox.com/")
time.sleep(2)

#opening Jobs option
search_job=driver.find_element_by_xpath("//a[@title='Company Salaries']")
search_job.click()
time.sleep(1)
#searching Data Scientist  jobs
search_njob=driver.find_element_by_id('jobProfileSearchbox')
search_njob.send_keys("Data Scientist") 
time.sleep(1)

data_select= driver.find_element_by_xpath("//div[@class='top-row']/p")
data_select.click()
time.sleep(1)
#company name
coy_tag=driver.find_elements_by_xpath("//div[@class='name']/a")
coy_list=[]
for c in coy_tag:
        company=c.text
        coy_list.append(company)

#total salary record
salary_tag=driver.find_elements_by_xpath("//div[@class='name']/span")
salary_list=[]
for s in salary_tag:
        salary=s.text
        salary_list.append(salary)
        
#average salary
avg_sal=driver.find_elements_by_xpath("//p[@class='averageCtc']")
avgs_list=[]
for a in avg_sal:
        avg_salary=a.text
        avgs_list.append(avg_salary)
        
#Minimum salary
minsal=driver.find_elements_by_xpath("//div[@class='value body-medium'][1]")
min_list=[]
for mini in minsal:
        minimum_s=mini.text
        min_list.append(minimum_s)
        
#maximum salary
maxsal=driver.find_elements_by_xpath("//div[@class='value body-medium'][2]")
max_list=[]
for maxi in maxsal:
        max_salary=maxi.text
        max_list.append(max_salary)
time.sleep(2)
#experience
exp_tag=driver.find_elements_by_xpath("//div[@class='salaries sbold-list-header'][1]")
exp_list=[]
for e in exp_tag:
        experience=e.text.split('\n . \n')[1]
        exp_list.append(experience)
        

abc=pd.DataFrame({})
abc['Company Name']=coy_list[0:10]
abc['Total salary record']=salary_list[0:10]
abc['Average Salary']=avgs_list[0:10]
abc['Min Salary']=min_list[0:10]
abc['Max Salary']=max_list[0:10]
abc['Experience']=exp_list[0:10]
```
